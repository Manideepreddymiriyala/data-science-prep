```python
import operator as op
from functools import reduce
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
%matplotlib inline
rcParams['figure.figsize'] = 15, 10
sns.set_style()
```

```python
def ncr(n: int, r: int) -> int:
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer / denom
```

```python
def probability_of_heads(num_flips: int, num_heads: int, p: float = 0.5) -> float:
    return ncr(num_flips, num_heads) * p**num_heads * (1 - p)**(num_flips - num_heads)
```

```python
heads = range(0, 1001)
p_of_heads = list(map(lambda x: probability_of_heads(1000, x), heads))
```

```python
plt.bar(heads, p_of_heads, color="#3D9970")
plt.title("PMF of $X \sim Binomial(p)$")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xlim(425, 575)
plt.grid()
```

```python
plt.bar(
    heads, 
    p_of_heads, 
    color=["#3D9970" if head < 550 else "#85144b" for head in heads], 
)
plt.title("PMF of $X \sim Binomial(p)$")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xlim(425, 575)
plt.grid()
```

```python
plt.bar(heads, p_of_heads, color=["#3D9970" if head < 550 else "#85144b" for head in heads])
plt.title("PMF of $X \sim Binomial(p)$ (Zoomed) -- Red area what we're interested in")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xlim(490, 570)
plt.grid()
```

```python
total_prob: float = 0
for i in range(550, 1001):
    total_prob += ncr(1000, i) * 0.5**1000
```
